## Analysis of the Perceptron Algorithm


```python
iters = []
seps = []
for sep in np.linspace(2, 10, 1000):
    X,y = Dataset(1000, n_dim=2, mode="L", w_1=np.array([1,2]), separability=sep, bias = 30).get_data()
    p = Perceptron(X, y)
    _,corr_count = p.train()
    iters.append(corr_count)
    seps.append(sep)

plt.plot(seps, iters)
plt.xlabel("Separability")
plt.ylabel("Iterations")
    
```
